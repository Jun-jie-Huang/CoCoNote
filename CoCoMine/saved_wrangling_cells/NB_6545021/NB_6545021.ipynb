```python
import numpy as np

import scipy

import pandas as pd 

from sklearn.linear_model import LinearRegression

import warnings 
 
warnings.filterwarnings('ignore')
```

```python
height = pd.read_csv('DATASET_NB_ROOT/NB_6545021/height_train.csv')

height.head(5)
```

```python
model = LinearRegression()

model.fit(X=height.loc[:,['father_height','mother_height']],y=height.child_height)
```

```python
print (model.coef_)
print ()
print (model.intercept_)
```

含义，在其他条件不变的情况下，父亲每升高1m，孩子升高0.25m 同理，母亲每升高1m，孩子升高0.26m
